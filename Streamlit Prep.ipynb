```python
vhost = pd.read_csv("C:\\Users\\hbalcera\\00 Eskwelabs\\Capstone\\clusters.csv", usecols=['Species','Host','Virus Clusters'])
vv = vhost.groupby(['Virus Clusters'])['Host'].apply(','.join)
vv = pd.DataFrame(vv)
```

```python
def deldups(string):
    res = []
    foo = string.split(',')
    for j in foo:
        if j not in res:
            res.append(j)
    val = ', '.join(res)
    return val
```

```python
hosts_avail = []
for i in range(len(hosts)):
    w=0
    x=0
    y=0 
    z=0
    host_i=''
    if hosts['Virulent_Counts'].iloc[i] > 1.0:
        for i in range(len(vhost[vhost['Virus Clusters']==0])):
            host_i += vhost[vhost['Host']].iloc[i]
            x = deldups(host_i)
    hosts_avail.append(x)
    if hosts['Infectious_Counts'].iloc[i] > 1.0:
        for i in range(len(vhost[vhost['Virus Clusters']==1])):
            host_i += vhost[vhost['Host']].iloc[i]
            y = deldups(host_i)
    hosts_avail.append(y)
    if hosts['Replicative_Counts'].iloc[i] > 1.0:
        for i in range(len(vhost[vhost['Virus Clusters']==2])):
            host_i += vhost[vhost['Host']].iloc[i]
            z = deldups(host_i)
    hosts_avail.append(z)
    if hosts['Replicative_Counts'].iloc[i] > 1.0:
        for i in range(len(vhost[vhost['Virus Clusters']==3])):
            host_i += vhost[vhost['Host']].iloc[i]
            w = deldups(host_i)
    hosts_avail.append(w)
```
